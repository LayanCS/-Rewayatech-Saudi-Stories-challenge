# Visualisation 

```python
import numpy
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import arabic_reshaper
%matplotlib inline
```

```python
df_love = pd.read_csv('love.csv')
df_hate = pd.read_csv('hate.csv')
df_count = pd.read_csv('countries_and_cities.csv')
df_place = pd.read_csv('Places.csv')
```

# 1: countries and cities

```python
x=[]
for locations in df_count.locations:
    x.append(arabic_reshaper.reshape(locations)[::-1])
```

```python
# For quantative and qualtative data the violin and box plot are the most suitable, but i am going to use the bar plot since it is easier in the summary reading
#Plot
fig, ax = plt.subplots(nrows=1, figsize = [7,5])
base_color = sb.color_palette()[0]

sb.barplot(data = df_count, x =x, y = 'freq', color = base_color, errwidth=0)


# Lables
title = 'تكرارالدول والمدن'
plt.title(arabic_reshaper.reshape(title)[::-1],fontsize=20)
plt.xlabel(arabic_reshaper.reshape('الدول والمدن')[::-1],fontsize=15)
plt.ylabel(arabic_reshaper.reshape('التكرار')[::-1],fontsize=15)
plt.tick_params(labelsize=13)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.savefig('المدن والمناطق.png', format='png', dpi=150, bbox_inches='tight')

arabic_reshaper.reshape(plt.show())
```
